```python
# Completamos DataSet
data_1['OperadOR'].replace(['?'],['Sin Datos'], inplace=True)
data_1['flight_no'].replace(['?'],['Sin Datos'], inplace=True)
data_1['route'].replace(['?'],['Sin Datos'], inplace=True)
data_1['ac_type'].replace(['?'],['Sin Datos'], inplace=True)
data_1['registration'].replace(['?'],['Sin Datos'], inplace=True)
data_1['cn_ln'].replace(['?'],['Sin Datos'], inplace=True)
data_1['summary'].replace(['?'],['Sin Datos'], inplace=True)
```

```python
# Completamos DataSet

data_1['all_aboard'].replace(['?'],['0'], inplace=True)	
data_1['PASAJEROS A BORDO'].replace(['?'],['0'], inplace=True)	
data_1['cantidad de fallecidos'].replace(['?'],['0'], inplace=True)	
data_1['crew_aboard'].replace(['?'],['0'], inplace=True)	
data_1['passenger_fatalities'].replace(['?'],['0'], inplace=True)	
data_1['crew_fatalities'].replace(['?'],['0'], inplace=True)	
data_1['ground'].replace(['?'],['0'], inplace=True)	
```

```python
# Modifica DataSet Columna Hora
data_1['HORA declarada'] = data_1['HORA declarada'].str.split('c ').str[-1]
data_1['HORA declarada'] = data_1['HORA declarada'].str.split('c: ').str[-1]
data_1['HORA declarada'] = data_1['HORA declarada'].str.split('c:').str[-1]
data_1['HORA declarada'] = data_1['HORA declarada'].str.split('Z').str[-1]
data_1['HORA declarada'] = data_1['HORA declarada'].replace(['?'],['00:00'])
data_1['HORA declarada'] = data_1['HORA declarada'].replace([''],['00:00'])
```

```python
# Funcion : para los datos faltante

def normal_horax(value):
    if ':' not in value:
        return f'{value[:2]}:{value[-2:]}'
    else:
        return value

data_1['HORA declarada']= data_1['HORA declarada'].apply(lambda x: normal_horax(x))
data_1['HORA declarada']
```

```python
## TRANFORMAR DATOS SIN 0 a 4 Zero

data_1['HORA declarada'] = data_1['HORA declarada'].astype(str)

# CANTIDAD DE 0
width = 5
  
# PRODUCTO MODIFICA LA CANTIDAD DE 0
data_1['HORA declarada']= data_1['HORA declarada'].str.zfill(width)
  
# SALIDA
data_1
```

```python
# Hora mal Escrita 

data_1['HORA declarada'] = data_1['HORA declarada'].replace(['90:00'],['09:00'])
data_1['HORA declarada'] = data_1['HORA declarada'].replace(['000:7'],['07:00'])
data_1['HORA declarada'] = data_1['HORA declarada'].replace(['24:45'],['02:45'])
```
